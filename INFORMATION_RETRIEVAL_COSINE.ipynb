```python
import re
def clean_tweets(review):
  cleaned_content = re.sub(r'https?:/\/\S+', '', review)
  cleaned_content = re.sub(r'@[A-Za-z0-9]+', '', cleaned_content)
  cleaned_content= re.sub('[^a-zA-Z0-9\.]', ' ', cleaned_content)
  cleaned_content = re.sub('[0-9]+','', cleaned_content)
  cleaned_content=" ".join(cleaned_content.split())
  cleaned_content=cleaned_content.lower()
  return cleaned_content

data['clean_review'] = data['review'].apply(clean_tweets)
```

```python
data['review'].iloc[0]
```

```python
documents=[]
documents=data['clean_review'].tolist()
documents[0]
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
new_docs=[]
for i in documents:
  tokenizer = nltk.RegexpTokenizer(r"[a-z,A-Z]+")
  new_tokens = tokenizer.tokenize(i)
  new_tokens = [t.lower() for t in new_tokens]
  new_tokens =[t for t in new_tokens if len(t)!=1]
  new_tokens =[t for t in new_tokens if len(t)!=2]
  new_tokens =[t for t in new_tokens if t not in stopwords.words("english")]
  new_review=""
  for j in new_tokens:
    new_review=new_review+' '+j
  new_docs.append(new_review) 
new_docs[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector_tfidf=TfidfVectorizer()
x=vector_tfidf.fit_transform(new_docs)
print("Shape of the vector is :",x.shape)
print("\nTfidf Vector :\n")
print(x.toarray())
```

```python
tfidf_dataframe=pd.DataFrame(x.toarray(),columns=vector_tfidf.get_feature_names())
tfidf_dataframe
```

```python
column_names=tfidf_dataframe.columns.tolist()
print(column_names)
```

```python
query_tfidf=vector_tfidf.transform([query])
print(query_tfidf.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim1=cosine_similarity(x,query_tfidf)
column_name=[]
column_name.append("similarity")
similarity_dataframe=pd.DataFrame(sim1,columns=column_name,index=new_docs)
# similarity_dataframe.rename(columns = {"0":'similarity'}, inplace = True)
similarity_dataframe
```
